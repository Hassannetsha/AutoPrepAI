```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer, KNNImputer
import glob
import os
```

```python
#get a list of file paths to all .csv files it finds.
path = os.path.expanduser("Input/Dataset_v2.csv") 

# glob.glob now looks for the file directly inside AutoPrepAI
csv_files = glob.glob(path)
# csv_files = glob.glob("/Graduation project/AutoPrepAI/*.csv", recursive=True)
if not csv_files:
    raise Exception("no CSV found under //input. add data and rerun")
DATA_PATH = csv_files[0] #picks the first one to work with
print("Using dataset:", DATA_PATH)
dataframe = pd.read_csv(DATA_PATH)
dataframe.head()   #shows the first 5 rows
```

```python
#data size(rows,coloumn)
print("Shape:", dataframe.shape)
#check for missing value and return set with same size but true in missing values else false ;isna;
print("\nMissing values per column:\n", dataframe.isna().sum())
```

```python
numerical_cols = dataframe.select_dtypes(include=[np.number]).columns.tolist()
non_numerical = [c for c in dataframe.columns if c not in numerical_cols]
print("The numerical columns:", numerical_cols)
print("the categorical columns:", non_numerical)
```

```python
if numerical_cols:
    imputer_mean = SimpleImputer(strategy="mean")
    dataframe_mean = dataframe.copy() #copy of dataframe to leave the original one unchanged
    dataframe_mean[numerical_cols] = imputer_mean.fit_transform(dataframe_mean[numerical_cols])
    print("After Mean imputation, missing values (in numeric coloumns):", dataframe_mean[numerical_cols].isna().sum().sum())
    display(dataframe_mean)
else:
    print("not a numeric column.")
```

```python
if numerical_cols:
    imputer_median = SimpleImputer(strategy="median")
    dataframe_median = dataframe.copy()
    dataframe_median[numerical_cols] = imputer_median.fit_transform(dataframe_median[numerical_cols])
    print("After Median imputation, missing values (in numeric coloumns):", dataframe_median[numerical_cols].isna().sum().sum())
    display(dataframe_median)
else:
    print("not a numeric columns")
```

```python
imputer_mostfrequent = SimpleImputer(strategy="most_frequent")
dataframe_mostfrequent = dataframe.copy()
cols = numerical_cols + non_numerical
if cols:
    dataframe_mostfrequent[cols] = imputer_mostfrequent.fit_transform(dataframe_mostfrequent[cols])
    print("After Most Frequent imputation, missing values (for all coloumns):", dataframe_mostfrequent.isna().sum().sum())
    display(dataframe_mostfrequent)
else:
    print("please add some coloumns")
```

```python
imputer_const = SimpleImputer(strategy="constant", fill_value="missing")
dataframe_const = dataframe.copy()
cols = numerical_cols + non_numerical
if cols:
    dataframe_const[cols] = imputer_const.fit_transform(dataframe_const[cols])
    print("After Constant imputation, missing values:", dataframe_const.isna().sum().sum())
    display(dataframe_const)
else:
    print("No columns to impute.")
```

```python
if numerical_cols:
    from sklearn.impute import KNNImputer
    #fill missing with avg of nearest 3 neighbors
    imputer_knn = KNNImputer(n_neighbors=3)
    dataframe_knn = dataframe.copy()
    dataframe_knn[numerical_cols] = imputer_knn.fit_transform(dataframe_knn[numerical_cols])
    print("After KNN imputation, missing values (numeric):", dataframe_knn[numerical_cols].isna().sum().sum())
    display(dataframe_knn)
else:
    print("nor numeric columns")
```

```python
print("Before: \n",dataframe)
print("\n After Mean imputation: \n", dataframe_mean if 'dataframe_mean' in globals() else "N/A")
print("\n After Median imputation: \n", dataframe_median if 'dataframe_median' in globals() else "N/A")
print("\n After Most Frequency imputation: \n", dataframe_mostfrequent if 'dataframe_mostfrequent' in globals() else "N/A")
print("\n After Constant imputation: \n", dataframe_const if 'dataframe_const' in globals() else "N/A")
print("\n After KNN imputation: \n", dataframe_knn if 'dataframe_knn' in globals() else "N/A")
```
